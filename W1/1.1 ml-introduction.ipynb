### Lecture 1.2

`Q. How an image (B&W specipiaclly) be represented to computer using machine language?`
>A. Each image is like a matrix of `m X n` where each data point is a pixel on the image. Each data point consists of a number between [0,1], depending on the shade of that pixel.
>- 0 : pure white
>- 1 : pure black
>- 0<1(floats) : shades of gray
>
>![](https://www.researchgate.net/publication/353889980/figure/fig1/AS:1056036085583872@1628790031292/Matrix-representation-of-a-8-bits-black-and-white-image-The-values-range-from-0-to-255.ppm)
>
>ML can do many tasks that human do well (e.g. recognition of hand-written digits), but are unable to elaborate the step by step preocess that they follow for solving the tasks

`Q. How can we as a human can recognise any handwritten digits with such accuracy?`
>A. In short, because we are trained to do so from the primary level of education. We have been seen those digits over and over from
>various sources and in various patterns, so much so, that we hardly make any mistake upon recognising digits or any symbol, in general.
>
>Similar approach is taken when building a ML model for any particular task. TRAINING DATA is provided to the model which is being
>learnt to recognise any relation or pattern b/w given images(input) and labels(output).

### Model: Mathematical relation
>building a model is nothing but to establish a formula b/w pixels in the image and labels.
>
>The simplest of them is a linear relationship.
$$
W0 + \begin{bmatrix}
W1 & W2 & . & . & . & Wn
\end{bmatrix} * \begin{bmatrix}
px1\\
px2\\
.\\
.\\
.\\
pxn
\end{bmatrix} = label(i)
$$
