### Lecture 1.2 : ML vs Traditional Programming

`Q. How an image (B&W specipiaclly) be represented to computer using machine language?`
>A. Each image is like a matrix of `m X n` where each data point is a pixel on the image. Each data point consists of a number between [0,1], depending on the shade of that pixel.
>- 0 : pure white
>- 1 : pure black
>- 0<1(floats) : shades of gray
>

>ML can do many tasks that human do well (e.g. recognition of hand-written digits), but are unable to elaborate the step by step preocess that they follow for solving the tasks

#### Model: Mathematical relation
>building a model is nothing but to establish a formula b/w pixels in the image and labels.
>
>The simplest of them is a linear relationship.
$$
W0 + \begin{bmatrix}
W1 & W2 & . & . & . & Wn
\end{bmatrix} * \begin{bmatrix}
px1\\
px2\\
.\\
.\\
.\\
pxn
\end{bmatrix} = label(i)
$$
>Our job is to find the weights parameters ($W1,W2,...Wn$)

#### Loss Function: the beacon of optimization
>In the process of optimization of parameters,. we initially give some random values to them. Then we measure the error of calculation
>b/w our calculated value and actual label. This error is known as `Loss Function`.
>
>By using standard optimization precude we try to minimize the Loss function and reach the desired output. These weights parameters we get, when Loss function is at its minimum is called as Optimal parameters.

`Q. What is Trained Model?`
>A. The final product/model we get through optimization procedures.

`Q. What is evalustion of ML Model?`
>A. Finding performance of model on test data based on some pre-decided metric

`Q. What is the Fundamental assumption of ML?`
>The Training and Test data would come from the same distribution
